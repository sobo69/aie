```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("adult.csv")
df.head()

print("Dataset Shape:", df.shape)
print("\nColumns:\n", df.columns)
df.replace('?', pd.NA, inplace=True)
df.dropna(inplace=True)
df.describe()

df.dtypes

print("Workclass Distribution:\n", df['workclass'].value_counts(), "\n")
print("Education Distribution:\n", df['education'].value_counts(), "\n")
print("Income Distribution:\n", df['income'].value_counts(), "\n")

plt.figure(figsize=(6,4))
df['age'].plot(kind='hist', bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
plt.figure(figsize=(6,4))
df['gender'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
print("Gender Counts:\n", df['gender'].value_counts())
plt.figure(figsize=(7,4))
df['race'].value_counts().plot(kind='bar')
plt.title("Race Distribution")
plt.xlabel("Race")
plt.ylabel("Count")
plt.show()
print("Race Counts:\n", df['race'].value_counts())


df['age_group'] = pd.cut(df['age'], bins=[0, 25, 35, 45, 55, 65, 100],
                         labels=['0-25','26-35','36-45','46-55','56-65','65+'])
plt.figure(figsize=(7,4))
df['age_group'].value_counts().sort_index().plot(kind='bar')
plt.title("Age Group Distribution")
plt.xlabel("Age Groups")
plt.ylabel("Count")
plt.show()
print("Age Group Counts:\n", df['age_group'].value_counts().sort_index())


print("DEMOGRAPHIC IMBALANCE SUMMARY")
print("\nGender Imbalance:")
print(df['gender'].value_counts(normalize=True) * 100)
print("\nRace Imbalance:")
print(df['race'].value_counts(normalize=True) * 100)
print("\nAge Group Imbalance:")
print(df['age_group'].value_counts(normalize=True) * 100)
```
